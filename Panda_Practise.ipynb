```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['diamond_clarity'], bins=20)
plt.xlabel('diamond_clarity')
plt.ylabel('frequency')
plt.title('Counts of Daimond')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['diamond_weight'], bins=20)
plt.xlabel('diamond_weight')
plt.ylabel('frequency')
plt.title('Distribution of Daimond Weight')
plt.show()
```

```python
plt.figure(figsize=(8 ,6))
df['diamond_color'].value_counts().plot(kind='bar')
plt.xlabel('country')
plt.ylabel('count')
plt.title('No. of Country')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['diamond_weight'], df['price'])
plt.xlabel('diamond_weight')
plt.ylabel('price')
plt.title('Daimond Weight VS Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['diamond_weight'], bins=20)
plt.xlabel('diamond Weight')
plt.ylabel('frequency')
plt.title('Fq. of Daimond Weight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['diamond_color'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie of Daimond Color')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df_grouped = df.groupby(['diamond_shape', 'diamond_cut_grade']).size().unstack()
df_grouped.plot(kind='bar', stacked=True)
plt.xlabel('daimond shape')
plt.ylabel('count')
plt.title('comparison of diamond shape & cut grade')
plt.legend(title='daimond cut grade')
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='diamond_cut_grade', y='price', data=df)
plt.xlabel('diamond cut grade')
plt.ylabel('price')
plt.title('comparison of diamond cut grade')
plt.show()
```

```python
numerical_cols = ['diamond_weight', 'price']
corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(8, 9))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
numerical_cols = ['diamond_weight', 'price']
sns.pairplot(df[numerical_cols])
plt.tiltle('Pair Plot of Numericals Column')
plt.show()
```
