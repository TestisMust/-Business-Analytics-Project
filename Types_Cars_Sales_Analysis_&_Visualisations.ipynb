# **Performing Visualisations**

---

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car
```

#**Type of Car Sold (Histogram)**

```python
plt.figure(figsize=(8,6))
plt.hist(car['transmission'], bins=20)
plt.xlabel('Car Type')
plt.ylabel('Counts')
plt.title('Type of Car Sold')
plt.show()
```

# **Maximum Car Sold in Year (Histogram)**

```python
plt.figure(figsize=(8,6))
plt.hist(car['year'], bins=20)
plt.xlabel('Year')
plt.ylabel('Counts')
plt.title('Years Count')
plt.show()
```

#**Maximum fuel used type (Histogram)**

```python
plt.figure(figsize=(8, 6))
car['fuelType'].value_counts().plot(kind='bar')
plt.xlabel('FuelType')
plt.ylabel('Numbers')
plt.title('Numbers of Fuel Type')
plt.show()
```

#**Measuring Mileage (Scatter)**

```python
plt.figure(figsize=(8, 6))
plt.scatter(car['mileage'], car['engineSize'])
plt.xlabel('mileage')
plt.ylabel('engineSize')
plt.title('Measuring Milage A/C to engine Size')
plt.show()
```

#**Car Type (Pie Chart)**

```python
plt.figure(figsize=(8, 6))
car['Make'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.title('CAR TYPE')
plt.show()
```

#**Maximum Type of Cars Transmission Sold (Bar)**

```python
plt.figure(figsize=(8, 6))
car_grouped = car.groupby(['transmission', 'Make']).size().unstack()
car_grouped.plot(kind="bar", stacked=True)
plt.xlabel('MODEL')
plt.ylabel('numbers')
plt.legend(title='CAR TYPE')
plt.show()
```

#**Comparison Model VS Price (Box Plot)**

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Make', y='price', data=car)
plt.xlabel('Car Model')
plt.ylabel('Price')
plt.title('comparison Model VS Price')
plt.show()
```

#**Correlation Between Tax & Price (HeatMap)**

```python
numerical_cols = ['tax', 'price']
corr_matrix = car[numerical_cols].corr()

plt.figure(figsize=(8, 9))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('correlation matrix')
plt.show()
```

# **Pair Plot of Mileage & Enginesize**

```python
numerical_cols = ['mileage', 'engineSize']
sns.pairplot(car[numerical_cols])
plt.title('Pair Plot')
plt.show()
```
